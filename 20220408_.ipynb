```python
import requests
from bs4 import BeautifulSoup

request = requests.get("http://www.cu.ac.kr/plaza/notice/notice")

html = request.text

soup = BeautifulSoup(html, 'html.parser')

trs = soup.select('div.board_list tr')
print(len(trs))

board_data = []

for tr in trs:
    tds = tr.find_all('td')
    
    if len(tds) == 0:
        continue
    board_data.append(
        [tds[1].text.strip(),
        tds[3].text.strip(),
        tds[4].text.strip(),
        tds[4].text.strip()])

import pandas as pd
board_df = pd.DataFrame(board_data, columns=['title', 'writer', 'data', 'views'])
board_df
```
